```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Dados em CSV

```python
df = pd.read_csv("DataTickets.csv")
print(df.shape)
df.head(12)
```

### Transformar do dados em um formato mais adequado apara gerar os gráficos

```python
# Definir um tema 'limpo'
sns.set_theme() 

# Aumentar a área do Gráfico
plt.figure(figsize=(16,12),facecolor='skyblue') 

# Barplot
ax = sns.barplot(x="Mes", y="Value", hue = "Status", data=df)

# adiciona valores no alto das barras
for container in ax.containers: 
    ax.bar_label(container, size = 18)
 
# Define o intervalo de valores do eixo Y
ax.set(ylim=(90, 250)) 

# Aplica o formato de três valores decimais no eixo 
from matplotlib.ticker import StrMethodFormatter # Biblioteca para formatar números nos eixos dos gráficos
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.3f}')) 

# Nomes e fontes dos eixos
plt.xlabel("Mês", size = 15) # Label e fontSize to eixo x
plt.ylabel("Quantidade de Tickets", size = 20) # Label e fontSize to eixo y
plt.title("Tendência dos Tickets", fontsize=20) # Título e fontSize do Gráfico
plt.yticks(size = 15)  # FontSize do eixo y
plt.xticks(size = 20, rotation=45) # FontSize do eixo x

# Nomes, fontes e cores da legenda
frame = plt.legend(title = "Status", fontsize=20, title_fontsize=25, bbox_to_anchor=(1.05, 1), borderaxespad=0).get_frame() # Box da legenda
frame.set_facecolor('r') # Cor do box da legenda

# Finaliza o Gráfico
plt.show()
```

## O Gráfico de barras é o mais apropriado para este contexto?

Nossos os olhos leem gráfico de barras facilmente. Eles comparam os pontos extremos as barras, de modo que é fácil ver rapidamente qual é a categoria maior, qual é a menor e a diferença incremental entre as categorias. (Kaflic,2019)

O gráfico de barras são comuns e fácil de ler, em vez de usar o poder do cérebro para tentar ler o gráfico o público utiliza para decidir qual informação extraír da apresentação.(Kaflic,2019)

## O Gráfico de linhas

Os gráficos de linhas são mais usados para registrar dados contínuos. Como, os dados estão fisicamente conectados por meio da linha, isso indica uma conexão entre eles, que pode não fazer sentido para dados categóricos. Frequentemente nossos dados contínuos estão em alguma unidade de tempo dias meses trimestre ou anos. (kaflic,2019)

Os gráficos de linha também podem ser usados para mostrar uma evolução temporal de uma variável.

```python
sns.set_theme(style="whitegrid") # fundo branco
plt.figure(figsize=(16,12)) # Aumentar a área do Gráfico

# ********
# Gráfico de linhas
ax = sns.lineplot(x="Mes", y="Value", hue = "Status", data=df, style='Status', markers=True, dashes=False, linewidth = 3, markersize = 15) # lineplot
# ********

plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.3f}')) # three decimal places
for x, y in zip(df['Mes'], df['Value']):
    ax.text(x = x,  
            y = y,  
            s = '{:.0f}'.format(y),  
            fontsize = 15
           ) 
ax.set(ylim=(0, 250)) # intervalo de valores do eixo Y

plt.xlabel("Mês", size = 15)
plt.ylabel("Quantidade de Tickets", size = 20)
plt.title("Tendência dos Tickets", fontsize=20)
plt.yticks(size = 15)  # FontSize do eixo y
plt.xticks(size = 20, rotation=45) # FontSize do eixo x
plt.legend(fontsize=20, bbox_to_anchor=(0.5, -0.25), loc = "lower center", ncol = df.shape[1], borderaxespad=0, frameon=False) # Box da legenda

plt.show()
```
