# Density-Based Clustering

## Objetivo

*   Aprender a usar o algoritmo do DBSCAN do Scikit-learn


```python
import numpy as np 
import pandas as pd
from sklearn.cluster import DBSCAN 
from sklearn.datasets import make_blobs 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
%matplotlib inline

np.random.seed(0)
```

## Gerar alguns dados 
### "Fazer uns Blobs"

```python
centroids = [[7,10], [0, 0], [5, -5], [-5, 7], [7,0]]

X, y = make_blobs(n_samples=150, centers=centroids, cluster_std=0.8)
plt.scatter(X[:, 0], X[:, 1], marker='.')
```

```python
X = StandardScaler().fit_transform(X)
plt.scatter(X[:, 0], X[:, 1], marker='.')
```

## DBSCAN
Nós Usaremos 2 parâmetros para o DBSCAN:

* Epsilon - Medida do Raio
* minimumSamples - quantos Vizinhos por raio



```python
epsilon = 0.2
minimumSamples = 6

db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)

labels = db.labels_
df = pd.DataFrame(labels)
UniqueLabels = df[0].unique()

UniqueLabels
```

```python
plt.figure(figsize=(18, 12))
for i in UniqueLabels:
    if i != -1:
        plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i)
    else: # Print OUTLIERS
        plt.scatter(X[labels == i , 0] , X[labels == i , 1] , label = i, marker="x", s=300, color = 'k')
        
plt.legend()
plt.show()
```
